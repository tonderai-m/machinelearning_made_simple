```python
plt.plot(x,x**2 + 3, "1--", x, x**2 + 15, ".--", x, -x**2, x, 2**x)
plt.xlim([-10,10])
plt.ylim([-50,150])
plt.ylabel("f(x)")
plt.xlabel("x")
plt.legend(['x**2 + 3','x**2 + 15','-x**2','2**x'], loc='lower center')
plt.grid(True)
plt.show()
```

Mixing line styles and seeing what to do after the fact mix and match

```python
plt.plot(x,x**2 + 3, "1--k", x, x**2 + 15, "b.--", x, -x**2, x, 2**x,'*r-.')
plt.xlim([-10,10])
plt.ylim([-50,150])
plt.ylabel("f(x)")
plt.xlabel("x")
plt.legend(['x**2 + 3','x**2 + 15','-x**2','2**x'], loc='lower center')
plt.grid(True)
plt.show()
```

```python
fig = plt.figure(constrained_layout=True)
gs = fig.add_gridspec(3, 3)
fig.add_subplot(gs[0,0])
fig.add_subplot(gs[1,0])
fig.add_subplot(gs[:,2])
```
