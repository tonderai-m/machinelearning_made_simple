```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0.000001, 1, 1000)
Y = np.log(X)
fig, ax = plt.subplots()
ax.plot(X, Y, color="green")
ax.set_ylabel("log x")
ax.set_xlabel("x")
plt.fill_between(
    x=X,
    y1=Y,
    where=Y < X,
    color="blue",
    alpha=0.2
)
```
