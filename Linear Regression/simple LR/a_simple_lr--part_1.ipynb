# In this tutorial we are going derive linear regression y = mx + c
* First we just plot a simple linear line 
* Write the equations to determine simple linear regression



```python
import numpy as np
import matplotlib.pyplot as plt
```

### y = 4x + 3

```python
m, c = 2, 3
```

```python


x = np.linspace(-10, 10, 400)  # 400 points between -10 and 10

# Compute y values
y = m * x + c

# Optionally add some noise
e = np.random.normal(0, 1, len(x)) # Gaussian noise
y1 = y + e
```

```python

# Plot the data
plt.plot(x, y, 'r-', label='Without Noise')
plt.scatter(x, y1, s=5, label='With Noise')
plt.legend()
plt.show()
```
